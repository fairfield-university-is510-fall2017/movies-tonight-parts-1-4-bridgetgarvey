## Normalization Analysis


### 1NF Conversion
The flaw of Format 3 is that there is no primary key designated for the table because there are 13 duplicate rows.  We must correct this so that the table can be in 1NF. 

To correct: Designate a primary and eliminate duplicate rows. 

TABLE(TName,Location,Phone,Mtitle,Showtime,Rating,Ccode,Cname)

### 2NF Conversion

determinant columns ---> dependent columns

TName--->(Location,Phone)

Mtitle--->(Rating)

THEATERS(TName,Location,Phone)

MOVIES(Mtitle,Rating)

CREDITS(Mtitle,Ccode,Cname)

SHOWS(TName,Mtitle,Showtime)

ARTISTS(ArtistID,Name)

### 3NF Conversion

There is a transitive dependency in Location--->(Phone)

determinant columns ---> dependent columns

THEATERS(TheaterID,Location,Phone)

MOVIES(MovieID,Rating)

CREDITS(MovieID,Ccode,ArtistID)

SHOWS(TheaterID,MovieID,Showtime)

ARTISTS(ArtistID,Name)
